```python
import numpy as np
import pandas as pd
import datetime
import pyfolio as pf
import matplotlib.pyplot as plt
import matplotlib
import seaborn as sns
import pytz
import empyrical
def local_csv(filename):
    return pd.read_csv(filename)
```

```python
# Global variables to tune the strategy
tolerance = 0.
look_back = 12

# Vol scaling
vol_flag = 1                  # Set flag to 1 for vol targeting
if vol_flag == 1:
    target_vol = 0.3
else:
    target_vol = 'no target vol'
```

# Implementing the Time Series Momentum strategy using Ex-Ante Volatility


```python

datasetSoFar = pd.read_csv("futuresTill2016.csv") 
# Setting the Date columns
datasetSoFar['Date'] = pd.to_datetime(datasetSoFar['Date'], format='%Y-%m-%d')
datasetSoFar.set_index('Date', inplace=True)
datasetSoFar
```

```python
standardIND = datasetSoFar.resample('BME').last().index
ithInd = pd.DataFrame(index=standardIND)
ithIndVol = pd.DataFrame(index=standardIND)
summStats = pd.DataFrame(index=['Asset', 'Start', 'Mean', 'Std', 'Sharpe Ratio'])

summStats = summStats.transpose()
listFutures = local_csv("futuresList.csv")
takeAll = summStats.reset_index().merge(listFutures)
takeAll.sort_values(by=["ASSET_CLASS", "FUTURES"], inplace=True)
del takeAll['Asset'], takeAll['index']
```

```python
for elem in datasetSoFar.columns:
    curr = datasetSoFar[elem]
    curr.dropna(inplace=True)
    
    dateStart = curr.index[0].strftime("%Y-%m-%d")   # store this to show when data series starts
    
    outInd = (1 + curr).cumprod()
    outInd.iloc[0] = 1
    
    # equation for ex ante vol estimate
    # .ewm is the exponentially weighted moving average
    dailyVolatility = curr.ewm(ignore_na=False,
                          adjust=True,
                          com=60,   
                          min_periods=0).std(bias=False)
    annualizedVol = dailyVolatility * np.sqrt(261) # annualise
    
    outInd = pd.concat([outInd, annualizedVol], axis=1)
    outInd.columns = [elem, 'vol']
    outIthInd = outInd.resample('BME').last().ffill()
    outIthInd.iloc[0, outIthInd.columns.get_loc(elem)] = 1
    ithInd = pd.concat([ithInd, outIthInd[elem]], axis=1)
    trash = outIthInd['vol']
    trash.name = elem + "_Vol"
    ithIndVol = pd.concat([ithIndVol, trash], axis=1)

    trashAvg = outInd[elem].pct_change().mean()*252
    trashDevn = outInd[elem].pct_change().std()*np.sqrt(252)
    sr = trashAvg / trashDevn
    contains = {'Asset': elem,
            'Start': dateStart,
            'Mean': np.round(trashAvg,4),
            'Std':  np.round(trashDevn,4),
            'Sharpe Ratio': np.round(sr,4),
            }
    summStats[elem] = pd.Series(contains)    
```

```python
summStats = summStats.transpose()
listFutures = local_csv("futuresList.csv")
takeAll = summStats.reset_index().merge(listFutures)
takeAll.sort_values(by=["ASSET_CLASS", "FUTURES"], inplace=True)
del takeAll['Asset'], takeAll['index']
```

Using the strategy mentioned in the paper which says the following :

=> Long those having positive momentum

=> Short those having negative momentum

```python
profNLoss = pd.DataFrame(index=standardIND)
lvg = pd.DataFrame(index=standardIND)
stratCummRetns = pd.DataFrame(index=standardIND)
```

```python
for elem in ithInd:
    dataFolio = pd.concat([ithInd[elem], ithIndVol[elem+"_Vol"]], axis=1)
    dataFolio['returns'] = dataFolio[elem].pct_change(look_back)
    
    dataFolio['pnl'] = 0.
    dataFolio['leverage'] = 0.
    try:
        for k, v in enumerate(dataFolio['returns']):
            if k <= look_back:
                # skip the first 12 observations as our first look back is supposed to be 12 and unkownn
                continue
            if dataFolio['returns'].iloc[k-1] < tolerance:
                # negative returns, sell and hold for 1 month, then close position
                if vol_flag == 1:
                    dataFolio['pnl'].iloc[k] = (dataFolio[elem].iloc[k - 1] / dataFolio[elem].iloc[k] - 1) * \
                                        target_vol / dataFolio[elem+"_Vol"].iloc[k - 1]
                    dataFolio['leverage'].iloc[k] = target_vol / dataFolio[elem+"_Vol"].iloc[k - 1]
                else:
                    dataFolio['pnl'].iloc[k] = (dataFolio[elem].iloc[k - 1] / dataFolio[elem].iloc[k] - 1)                   
                    dataFolio['leverage'].iloc[k] = 1.
            elif dataFolio['returns'].iloc[k-1] > tolerance:
                # positive returns, buy and hold for 1 month, then close position
                if vol_flag == 1:
                    dataFolio['pnl'].iloc[k] = (dataFolio[elem].iloc[k] / dataFolio[elem].iloc[k - 1] - 1) * \
                                        target_vol / dataFolio[elem+"_Vol"].iloc[k - 1]
                    dataFolio['leverage'].iloc[k] = target_vol / dataFolio[elem+"_Vol"].iloc[k - 1]
                else:
                    dataFolio['pnl'].iloc[k] = (dataFolio[elem].iloc[k] / dataFolio[elem].iloc[k - 1] - 1) 
                    dataFolio['leverage'].iloc[k] = 1.
    except: pass    
    # convert to cummulative index
    profNLoss = pd.concat([profNLoss, dataFolio['pnl']], axis=1)
    lvg = pd.concat([lvg, dataFolio['leverage']], axis=1)
    
    outInd = (1 + dataFolio['pnl'][13:]).cumprod()
    outInd[0] = 1
    stratCummRetns = pd.concat([stratCummRetns, outInd], axis=1) 
```

```python
profNLoss.columns = datasetSoFar.columns
lvg.columns = lvg.columns
stratCummRetns.columns = datasetSoFar.columns
dataFolio = profNLoss
dataFolio['port_avg'] = dataFolio.mean(skipna=True, axis=1)
Strategy = dataFolio['port_avg'].copy()
Strategy.name = "TSMOM with Vol"
dataport_index = (1 + dataFolio['port_avg']).cumprod()
```

```python


# Calculate cumulative returns
dataFolio['cumulative_returns'] = (1 + dataFolio['port_avg']).cumprod()

# Calculate drawdowns
dataFolio['peak'] = dataFolio['cumulative_returns'].cummax()
dataFolio['drawdown'] = dataFolio['cumulative_returns'] / dataFolio['peak'] - 1

# Find the top 5 drawdowns
top_drawdowns = dataFolio.nsmallest(5, 'drawdown')

# Plot
plt.figure(figsize=(14, 7))
plt.plot(dataFolio.index, dataFolio['cumulative_returns'], label='Cumulative Portfolio Returns', color='orange')

# Highlight top 5 drawdown periods
for start in top_drawdowns.index:
    end = dataFolio.index[dataFolio.index > start][dataFolio['cumulative_returns'][dataFolio.index > start] == dataFolio['cumulative_returns'][dataFolio.index > start].max()].min()
    plt.axvspan(start, end, color='red', alpha=0.3)

# Adding labels and title
plt.title('Cumulative Portfolio Returns with Top 5 Drawdown Periods Highlighted')
plt.xlabel('Date')
plt.ylabel('Cumulative Returns')
plt.legend()
plt.grid()
plt.show()
```
